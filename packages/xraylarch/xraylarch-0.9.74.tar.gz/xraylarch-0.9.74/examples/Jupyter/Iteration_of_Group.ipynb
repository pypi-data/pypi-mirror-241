# Iteration of larch Group class
This notebook provides a example of using larch Group in iterative method to deal with multiple data sets.

```python
# importing dependencies to run this notebook
from larch import Group
from larch.io import read_ascii, read_athena

import os
```

## Before we start: Group class now supports 2 ways of accessing the data
1. Using attribute name
2. Using dictionary key

These method will provide a more convenient way to access the data in the group.

```python
data = read_ascii('../xafsdata/fe.060')

# Print the data
data
```

```python
# Printing the content in the group via the attribute name

data.i0
```

```python
# Printing the content in the group via the key name

data['i0']
```

# Using regular Groups

```python
data = read_ascii('../xafsdata/fe.060')
```

```python
# Printing the content in the group

data
```

```python
# Iterating over the group
# the default iterator will iterate over the keys in the group
# keys can also be accessed using the keys() method

print("Iteration over the group will iterate over the keys in the group")
for key in data:
    print(key)

print("\n")

print("The keys can also be accessed using the keys() method")    
for key in data.keys():
    print(key)

```

```python
# Iterating over the group with the values

for value in data.values():
    print(value)
```

```python
# Iterating over the group with the items

for key, value in data.items():
    print(key, value)
```

# Iterating over Athena Groups

Athena groups read from Athena project files has a different structure of the Group class.
There for iteration of Athena Groups are modified to be iterate over subgroups of Athena groups.

```python
aprjfile = os.path.join("..", "pca", "cyanobacteria.prj")
aprj = read_athena(aprjfile)
```

```python
# Iteration of the group will have iterate over the keys in the group
# keys can also be accessed using the keys() method

print("Iteration over the group will iterate over the keys in the group")
for key in aprj:
    print(key)
    
print("\n")

print("The keys can also be accessed using the keys() method")
for key in aprj.keys():
    print(key)
    
```

```python
# Iterating over the group with the values
# the iteration over value will give a reference to the subgroups in the group

for value in aprj.values():
    print(value)
```

```python
# Iterating over the group with the items

for key, value in aprj.items():
    print(key, value)
```

# Application of the iteration method

Iteration over the groups will provide a easy way to deal with multiple data sets.

```python
# plotting all the subgroups in the athena project file

import matplotlib.pyplot as plt

for key, group in aprj.items():
    plt.plot(group.energy, group.mu, label=key)

plt.xlabel('Energy (eV)')
plt.ylabel('mu')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.plot()
```
